## Create a line plot showing the number of marriages and divorces per capita in theU.S. between 1867 and 2014. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
USM = pd.read_csv('us-marriages-divorces-1867-2014.csv')
USM.info()
```

```python
USM.head()
```

```python
years = USM['Year']
marriages = USM['Marriages_per_1000']
divorces = USM['Divorces_per_1000']
USM = plt.figure(figsize=(16,8))
USM = plt.plot(years, marriages, label='Marriages')
USM = plt.plot(years, divorces, label='Divorces')
USM = plt.title("Number of marriages and divorces per capita in the U.S. between 1867 and 2014")
USM = plt.xlabel("Years",fontsize=14)
USM = plt.legend(fontsize = 12, loc = "upper left")
USM = plt.ylabel("Marriages",fontsize=14)
USM = plt.grid(True)
USM
```

```python
Scplot = pd.read_csv("arcade-revenue-vs-cs-doctorates.csv")
Scplot
```

```python

# Plotting
plt.figure(figsize=(12, 7))
scatter = plt.scatter(Scplot['Total Arcade Revenue (billions)'], Scplot['Computer Science Doctorates Awarded (US)'], c=Scplot['Year'], cmap='viridis', s=100, edgecolors='k', alpha=0.7)

# Adding labels and title
plt.xlabel("Total Arcade Revenue (billions)")
plt.ylabel("Computer Science Doctorates Awarded (US)")
plt.title("Relationship between Arcade Revenue and CS Doctorates (2000-2009)")

# Adding colorbar
cbar = plt.colorbar(scatter)
cbar.set_label('Year')

# Show the plot
plt.show()
```
